#Functions in Python

```python
def PrintHello():
    print("Hello from the function printHello")

PrintHello()
```

```python
# Now i am going to discuss the argumented functions
# here i have the function called as "printHello"

def printHello(fname):
    print("Hello : ",fname)

printHello("hamza")
printHello("haider")
```

```python
# it is the two argumented functions

def printHello(fname, lname):
    print("Hello : ",fname, " ",lname)

printHello("hamza","Habib")
printHello("haider", "Habib")
```

```python
# if the number of the arguments are unknown so we write the "*" symbol in the argument section

def KidsNames(*names):
    print("The youngest child is : ",names[1])

KidsNames("hamza","Haider","Umar","Hassan")
```

```python
# now talking about the keyword arguments
def childs(child3, child2, child1):
    print("The youngest child is : ", child3)

childs(child1 = "hamza", child2 = "Nimra", child3 = "Hiader")
```

```python
# Default parameter value

def Country(country = "Pakistan"):
    print("My country name is : ", country)

Country()
Country("Afghanistan")
Country("Saudi Arab")
```

```python
# Passing list as an arguments

def FruitsList(food):
    for x in food:
        print(x)
fruits = ["apple", "banana", "cherry"]
FruitsList(fruits)
```
